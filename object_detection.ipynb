```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import cv2
```

```python
model= torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
model
```

```python
img='C:\\Users\\white\\Downloads\\yolov5_OB\\IMG-20230126-WA0037.jpg'
```

```python
results= model(img)
print(results)
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap= cv2.VideoCapture(0)
 
while cap.isOpened():
    ret,frame=cap.read()
    results=model(frame)
    cv2.imshow('YOLO',np.squeeze(results.render()))
    if cv2.waitKey(10) & 0xFF ==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
